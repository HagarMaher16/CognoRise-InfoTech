```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
from sklearn.preprocessing import LabelEncoder
```

# Data Prepration

```python
# Load Data
data = pd.read_csv('/kaggle/input/car-data/CarPrice_Assignment.csv')
data.head()
```

```python
# Prepare x and y
x = data.drop(['car_ID', 'price'], axis =1)
y = data['price']
print(x.head())
print(y)
```

```python
# Label Encoder for x
label_encoder = LabelEncoder()
x['CarName'] = label_encoder.fit_transform(x['CarName'])
x['fueltype'] = label_encoder.fit_transform(x['fueltype'])
x['aspiration'] = label_encoder.fit_transform(x['aspiration'])
x['doornumber'] = label_encoder.fit_transform(x['doornumber'])
x['carbody'] = label_encoder.fit_transform(x['carbody'])
x['drivewheel'] = label_encoder.fit_transform(x['drivewheel'])
x['enginelocation'] = label_encoder.fit_transform(x['enginelocation'])
x['cylindernumber'] = label_encoder.fit_transform(x['cylindernumber'])
x['fuelsystem'] = label_encoder.fit_transform(x['fuelsystem'])
x['enginetype'] = label_encoder.fit_transform(x['enginetype'])
```

# Linear Regression Model 

```python
# Fitting Linear Regression to the training set 
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(x, y)
```

```python
# Predicting the test set result  
y_pred = model.predict(x)
```

```python
from sklearn.metrics import mean_squared_error
print(f'MSE : {mean_squared_error(y, y_pred)}')
```
